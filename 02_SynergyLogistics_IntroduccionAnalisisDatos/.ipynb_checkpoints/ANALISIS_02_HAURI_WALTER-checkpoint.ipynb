# Proyecto 2: Introducción al análisis de datos
## Synergy Logistics

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Lectura de datos

```python
todos_datos = pd.read_csv("synergy_logistics_database.csv", encoding='utf-8', index_col="register_id", parse_dates=["year", "date"])
todos_datos.head()
```

```python
t_modes = todos_datos["transport_mode"].unique()
ordenar_modes = lambda df: [mode for mode in t_modes if mode in df["transport_mode"].unique()] # seleccionar los labels que sí estén disponibles, manteniendo el orden
```

### Rutas de importación y exportación
Agrupamos todo por rutas, para manejar los datos de forma simplificada

```python
agrupados_rutas = todos_datos.groupby(["origin", "destination", "transport_mode"]) # agrupamos por rutas, según origen, destino y tipo de transporte
descripcion_rutas = agrupados_rutas.agg({'total_value': ['sum','count']})["total_value"] # resumimos la información en cuántas veces aparece cada ruta, y cuánto genera cada una
```

### Rutas más demandadas

Imprimir las 10 rutas más demandadas (con más trayectos en la base de datos)

```python
mejores_rutas = descripcion_rutas.sort_values(by="count", ascending=False) # ordenamos por demanda
mejores10_rutas = mejores_rutas.head(10) # obtenemos las 10 rutas con más demanda
mejores10_rutas
```

```python
ax = sns.barplot(data=mejores10_rutas.reset_index().groupby(["transport_mode"]).count().reset_index(), x="count", y="transport_mode");
ax.set_xlabel("Frecuencia");
ax.set_ylabel("Medio de transporte");
ax.set_title("Medios de transporte para las 10 rutas más utilizadas");
```

#### **Conclusión:** De las 10 más utilizadas, 7 son marítimas lo que representa el 75% del total.

Distribución de ganancias por tipo de ruta en las 10 más demandadas

```python
m10_rutas = mejores10_rutas.reset_index()
plt.figure(figsize=(10, 10))
ax = sns.barplot(x="transport_mode", y="sum", data=m10_rutas, order=ordenar_modes(m10_rutas))
ax.set_xlabel("Medio de transporte");
ax.set_ylabel("Valor total por ruta");
ax.set_title("Valor total generado por ruta para las 10 mejores según el medio de transporte");
```

#### **Conclusión**: El valor total generado por ruta más alto corresponde a la única vía aérea, la cual es casi 4 veces mayor a la mediana de las marítimas o terrestres.

```python
suma10_rutas = mejores10_rutas["sum"].sum() # suma de valor total de las 10 mejores rutas
rutas_restantes = mejores_rutas.drop(mejores10_rutas.index)
suma_rutas_restantes = rutas_restantes["sum"].sum() # suma de valor total de las rutas restantes

print(f"El valor total de las 10 mejores rutas es: {suma10_rutas}\nEl valor total de las {rutas_restantes.shape[0]} rutas restantes es: {suma_rutas_restantes}\nEl porcentaje del total que representan las 10 mejores es: {suma10_rutas/suma_rutas_restantes*100}")
```

#### **Conclusión**: Las 10 mejores rutas generan casi el 25% del valor total, mientras el otro 75% es generado por las 192 rutas restantes.

## Medio de transporte utilizado

```python
agrupados_transporte = todos_datos.groupby(["transport_mode"]) # agrupamos por tipo de transporte
descripcion_transporte = agrupados_transporte.agg({'total_value': ['sum','count']})["total_value"] # resumimos la información en cuántas veces aparece cada ruta, y cuánto genera cada una
```

```python
descripcion_transporte.sort_values(by="count", ascending=False).head()
```

#### **Conclusión**: Por más de 3 veces, el medio de transporte de ruta más utilizado fue el marítimo, comparado con los demás.

Distribución de los valores totales por tipo de transporte utilizado

```python
plt.figure(figsize=(10, 10))
ax = sns.boxplot(data=todos_datos, x="transport_mode", y="total_value", hue="direction", order=ordenar_modes(todos_datos));
ax.set_xlabel("Medio de transporte");
ax.set_ylabel("Valor total por ruta");
ax.set_title("Distribución del valor total generado por ruta según medio de transporte y dirección");
# Cambiar el formato de la caja de leyenda
leg = ax.get_legend()
leg.set_title("Direction")
```

Ordenamos los outliers que no nos permiten visualizar de forma idónea los valores más representativos de la distribución

```python
k = 1.5 # Tukey's definition of outlier value
q1 = agrupados_transporte["total_value"].quantile(0.25) # primer cuartil
q3 = agrupados_transporte["total_value"].quantile(0.75) # tercer cuartil
outliers_maximos = q3 + k * (q3 - q1) # tomado de: https://en.wikipedia.org/wiki/Outlier
todos_sin_outliers = todos_datos[todos_datos["total_value"] <= outliers_maximos[todos_datos["transport_mode"]].tolist()] # truncamos (eliminamos) los outliers para que no aparezcan en la gráfica
```

```python
plt.figure(figsize=(10, 10))
ax = sns.boxplot(data=todos_sin_outliers, x="transport_mode", y="total_value", hue="direction", order=ordenar_modes(todos_sin_outliers));
ax.set_xlabel("Medio de transporte");
ax.set_ylabel("Valor total por ruta");
ax.set_title("Distribución del valor total generado por ruta según medio de transporte y dirección");
# Cambiar el formato de la caja de leyenda
leg = ax.get_legend()
leg.set_title("Direction")
```

#### **Conclusión**: Para todos los medios de transporte, la distribución del valor total obtenido por importaciones es más alto que el de exportaciones, con excepción de unos valores atípicos tanto en rutas marítimas como en caminos terrestres. A su vez, los valores totales obtenidos por ruta en transporte aéreo son mayores comparado con cualquiera de los demás medios.

Visualizamos las ganancias según tipo de transporte y las clasificamos por exportaciones e importaciones

```python
# to_ym = lambda a: a.strftime("%Y-%m")
# todos_datos["monthly"] = pd.to_datetime(pd.Series(map(to_ym, todos_datos["date"]))) # dummy way
todos_datos['year_month'] = pd.to_datetime(todos_datos['date'].dt.strftime('%Y-%m'))
```

```python
plt.figure(figsize=(10, 10))
ax = sns.relplot(data=todos_datos, x="year", y="total_value", hue="transport_mode", col="direction", kind="line");
ax.axes.flat[0].set_xlabel("Año");
ax.axes.flat[1].set_xlabel("Año");
ax.axes.flat[0].set_ylabel("Valor total por ruta");
```

#### **Conclusión**: Las tendencias de valor total por medio de transporte apuntan a que las exportaciones en general tienden a crecer con el paso de los años, particularmente las rutas aéreas, mientras que la caída de las rutas de caminos terrestres tienen una clara tendencia a disminuir. Por el lado de las importaciones, las tendencias son mayormente en aumento, a excepción de laaérea, que presenta una fuerte caída en el 2020. Los demás medios de transporte presentan una tendencia ligeramente marcada al aumento, destacando el repunte al final de las vías terrestres, tanto en vías férreas como en caminos.

Visualizamos el número de rutas activas a lo largo del tiempo dependiendo del tipo de transporte

```python
agrupados_mes = todos_datos.groupby(["year_month", "transport_mode", "direction"])
agrupados_mes = agrupados_mes.count().reset_index()
plt.figure(figsize=(10, 10))
ax = sns.relplot(data=agrupados_mes, x="year_month", y="total_value", hue="transport_mode", col="direction", style="transport_mode", kind="line");
ax.axes.flat[0].set_xlabel("Fecha");
ax.axes.flat[1].set_xlabel("Fecha");
ax.axes.flat[0].set_ylabel("Número de rutas");
```

```python
plt.figure(figsize=(10, 10))
agrupados_mes["month"] = agrupados_mes["year_month"].dt.month # seleccionamos únicamente el mes
ax = sns.catplot(data=agrupados_mes, x="month", y="total_value", col="direction", kind="bar");
ax.axes.flat[0].set_xlabel("Mes");
ax.axes.flat[1].set_xlabel("Mes");
ax.axes.flat[0].set_ylabel("Número de rutas");
```

#### **Conclusión**: Un patrón claro de estacionalidad se nota tanto en las importaciones como las exportaciones de forma anual, donde del mes de febrero a junio son los meses donde mayor demanda existe de todos los tipos de rutas. Para un análisis más detallado de patrones sería necesario un análisis con técnicas temporales que retiren la contribución estacional tan marcada. A su vez es evidente que el primer semestre del año constituye el momento crucial para el uso de las rutas, independientemente del medio de transporte.

## Valor total de importaciones y exportaciones

Buscamos las rutas que representan el 80% de las ganancias totales y las filtramos

```python
agrupados_paises = todos_datos.groupby(["origin", "destination", "direction"]) # agrupamos por origen, destino y tipo de ruta
mayores_ventas = agrupados_paises.sum().reset_index().sort_values(by="total_value", ascending=False) # ordenamos por valor de mayor a menor

total = mayores_ventas["total_value"].sum() # total de valor de todos los datos
mayores_ventas["cumsum"] = mayores_ventas["total_value"].cumsum() # suma acumulada de los valores totales

porcentaje = 0.8 # 80% del total de ganancias
porcentaje_total = mayores_ventas[mayores_ventas["cumsum"] <= total * porcentaje] # filtramos para quedarnos únicamente con las ventas que representaron menos o igual al porcentaje solicitado en la suma acumulada
porcentaje_total
```

### Países de origen con mayor aportación al valor y mayor número de rutas activas

```python
mejores_origenes = porcentaje_total.groupby("origin").agg({'total_value': ['sum','count']})["total_value"].sort_values(by="sum", ascending=False)
mejores_origenes
```

### Países destino con mayor aportación al valor y mayor número de rutas activas

```python
mejores_destinos = porcentaje_total.groupby("destination").agg({'total_value': ['sum','count']})["total_value"].sort_values(by="sum", ascending=False)
mejores_destinos
```

```python
mejores_origenes["total_values_per_route"] = mejores_origenes["sum"] / mejores_origenes["count"] # sacamos el valor total promedio por ruta para cada país
mejores_destinos["total_values_per_route"] = mejores_destinos["sum"] / mejores_destinos["count"] # sacamos el valor total promedio por ruta para cada país

plt.figure(figsize=(16, 16))
plt.subplot(211)
ax1 = sns.barplot(data=mejores_origenes, x="total_values_per_route", y="origin", order=mejores_origenes.sort_values(by="total_values_per_route")["origin"]);
ax1.set_xlabel("Valor total por ruta");
ax1.set_ylabel("País");
ax1.set_title("type = origin");
plt.subplot(212)
ax2 = sns.barplot(data=mejores_destinos, x="total_values_per_route", y="destination", order=mejores_destinos.sort_values(by="total_values_per_route")["destination"]);
ax2.set_xlabel("Valor total por ruta");
ax2.set_ylabel("País");
ax2.set_title("type = destination");
```

#### **Conclusión**: Las rutas comerciales con origen en países desarrollados son las que mayor aportación dan al valor total, tanto en número y ganancias netas, como por cada ruta, destacando China y USE en el primer caso, y Canadá en el segundo. Por parte de los destinos, son mayormente países en vías de desarrollo aquellos que predominan tanto en cantidad, como en valor total; destacando el caso de México, Vietnam y Tailandia.
